### 2. Interactive Time Series Visualization: Monthly Sales Trends
- **Task:** Build an interactive line chart displaying total Sales over time.  
- **Hint:** Convert **Order Date** into datetime format, aggregate sales by month, and use Plotly Express for interactivity.

```python
df['Order_Month_Num'] = df['Order Date'].dt.month
monthly_sales = df.groupby('Order_Month_Num')['Sales'].sum().reset_index()
sns.lineplot(data = monthly_sales , x="Order_Month_Num", y ="Sales")
plt.xticks(ticks=range(1,13), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
])
```

### 4. Box Plot: Sales Distribution by Sub-Category
